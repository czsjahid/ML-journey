```python
from sklearn import preprocessing
import numpy as np
```

```python
data = np.array( [[ -1, 2, 0],
                  [ 2, -1, 2],
                  [ 1, -1, -1]])
```

```python
data.mean( axis =0 )
```

```python
data = preprocessing.scale(data)
```

```python
data.mean( axis =0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data)
data_standard = scaler.transform(data)
```

```python
data_standard.mean( axis = 0)
```
