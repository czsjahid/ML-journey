**Reason: **After the shuffle, the old index values (e.g., 0, 1, 2, ...) are now scattered randomly next to their rows (e.g., 15, 78, 4, ...). This can be confusing and cause issues in later operations.

**.reset_index(drop=True):** This corrects the index:

The **drop=True** parameter ensures the old, scattered index values are discarded and do not become a redundant column in the new shuffled DataFrame.

```python
irisgroup = iris.groupby('species')
```

This command does not calculate any final metrics; it creates an intermediate **GroupBy object** and assigns it to the variable irisgroup.**Grouping: **It tells Pandas to split the iris DataFrame into separate logical groups based on the unique values found in the $\text{'species'}$ column (e.g., 'Setosa', 'Versicolor', 'Virginica').

**Preparation:** The irisgroup variable now holds the "recipe" for these groups, ready for a subsequent aggregation method (like $\text{.mean()}$, $\text{.max()}$, or $\text{.count()}$) to be called on it to produce meaningful results.

```python
sns.pairplot(iris, hue= 'species')
```

This command uses the Seaborn statistical data visualization library (aliased as $\text{sns}$) to create a Pair Plot of the iris DataFrame.

**Pair Plot Creation:** It generates a grid of scatter plots where every numerical column in the DataFrame is plotted against every other numerical column.

**Diagonal Plots:** The diagonal cells of the grid will show histograms or KDE (Kernel Density Estimates) for each individual feature.

**Color Coding:** The hue='species' parameter is crucial: it tells Seaborn to **color-code** all the plots based on the unique values in the $\text{'species'}$ column, allowing for visual inspection of how the different flower species separate based on their features.

**Kernel Density Estimation (KDE)** is a non-parametric way to estimate the **probability density function (PDF)** of a continuous random variable. Essentially, it's a technique used to visualize the distribution of a dataset by smoothing out the data points to create a continuous curve.

```python
sns.pairplot(iris)
```

```python
plt.figure( figsize = (12,8))
```

This command is used in conjunction with Matplotlib.pyplot (often aliased as $\text{plt}$) and serves to create a new figure object and define its size.

**Figure Creation:**  $\text{plt.figure()}$ is the function that initializes a new, blank canvas or container for your plot elements (axes, titles, labels, etc.).

**Size Definition:** **The figsize=(12, 8)** parameter sets the width and height of this canvas, measured in inches. In this case, the resulting plot area will be 12 inches wide and 8 inches tall.

```python
plt.subplot(2,2,1)
plt.hist(iris['sepal_length'], edgecolor = 'black', linewidth = 1.2)
```

This command is used in Matplotlib (aliased as $\text{plt}$) to divide the current figure (the canvas created by $\text{plt.figure()}$) into a grid of smaller plots and then select one of those positions to be the active drawing area.

**Grid Definition (2, 2):** The first two arguments, 2 and 2, define the layout of the grid: 2 rows and 2 columns. This means the figure will be divided into $2 \times 2 = 4$ total plotting areas.

**Position Selection (1):** The third argument, 1, specifies which of the subplots to select. The positions are numbered sequentially starting from 1 in the top-left corner, moving across the rows.

```python
plt.subplot(2,2,2)
plt.hist(iris['petal_length'], edgecolor = 'black', linewidth = 1.2)
plt.subplot(2,2,3)
plt.hist(iris['sepal_width'], edgecolor = 'black', linewidth = 1.2)
plt.subplot(2,2,4)
plt.hist(iris['petal_width'], edgecolor = 'black', linewidth = 1.2)
plt.show()
```

Rest three subplots are showing here.

```python
plt.figure( figsize = (12,8))
plt.subplot(2,2,1)
sns.violinplot(x = 'species',y=  'sepal_length', data = iris)
plt.subplot(2,2,2)
sns.violinplot(x = 'species', y= 'petal_length' , data = iris )
plt.subplot(2,2,3)
sns.violinplot(x = 'species', y= 'sepal_width', data = iris)
plt.subplot(2,2,4)
sns.violinplot( x = 'species', y=  'petal_width' , data = iris)
plt.show()
```

A violin plot is essentially a **mirrored KDE** plot turned on its side. Since the KDE itself represents the **probability density function (PDF)** of a continuous variable, it only needs one side (either the left or the right) to convey the data distribution's shape and spread.

The two symmetrical halves are mainly used for:

**Balance**: They make the plot visually balanced and resemble a violin (hence the name).

**Clarity**: The symmetrical shape clearly emphasizes the density function compared to a simple half-shape or a box plot alone.
