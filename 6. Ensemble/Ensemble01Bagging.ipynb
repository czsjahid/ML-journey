```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,roc_auc_score
```

```python
def print_score(clf, X_train, X_test, y_train, y_test, train=True):
  lb = preprocessing.LabelBinarizer()
  if train:
    y_pred = clf.predict(X_train)
    print("Train Result:\n")
    print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train, y_pred)))
    print("Classification Report:\n {} \n".format(classification_report(y_train, y_pred)))
    print("Confusion Matrix:\n {} \n".format(confusion_matrix(y_train, y_pred)))
    print("ROC AUC: {0:.4f}\n".format(roc_auc_score(lb.fit_transform(y_train), lb.transform(y_pred))))
    res = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
    print("Average Accuracy: \t {0:.4f}".format(np.mean(res)))
    print("Accuracy SD: \t\t {0:.4f}".format(np.std(res)))

  elif train == False: # Corrected the syntax error here
    # ...
    # test performance
    # ...
    res_test = clf.predict(X_test)
    lb.fit(y_train) # Fit the binarizer on the training data

    print("Test Result: \n")
    print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test,
                                                          res_test)))

    print("Classification Report: \n{}\n".format(classification_report(y_test,
                                                                      res_test)))

    print("Confusion Matrix: \n{}\n".format(confusion_matrix(y_test,
                                                           res_test)))

    print("ROC AUC: {0:.4f}\n".format(roc_auc_score(lb.transform(y_test),
                                                    lb.transform(res_test))))
```

```python
clf1 = DecisionTreeClassifier(random_state=42)
```

```python
clf1.fit( X_train, y_train)
```

```python
print_score(clf1, X_train, X_test, y_train, y_test, train=True)
```

```python
print_score(clf1, X_train, X_test, y_train, y_test, train=False)
```

```python
clf_bag = BaggingClassifier(estimator=clf1, n_estimators=1000, n_jobs=-1, random_state=42)
```

```python
clf_bag.fit( X_train, y_train)
```

```python
print_score(clf_bag, X_train, X_test, y_train, y_test, train=True)
```

```python
print_score(clf_bag, X_train, X_test, y_train, y_test, train=False)
```

```python
clf_bagOOB = BaggingClassifier(estimator=clf1, n_estimators=1000, n_jobs=-1, random_state=42, oob_score = True)
```

```python
clf_bagOOB.fit( X_train, y_train)
```

```python
print_score(clf_bagOOB, X_train, X_test, y_train, y_test, train=True)
```
