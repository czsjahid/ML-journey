```python
import numpy as np
```

```python
import tensorflow as tf
```

```python
imdb = tf.keras.datasets.imdb
```

```python
max_features = 5000
```

```python
(x_train, y_train), (x_test,y_test) = imdb.load_data(num_words=max_features)
```

```python
x_train
```

```python
dictionary = imdb.get_word_index()
```

```python
dictionary['man']
```

```python
def matrix_seq(data,dim=max_features):
  results=np.zeros((len(data),dim))
  for i,seq in enumerate(data):
    results[i,seq]=1
    return results
```

To prepare the data for a neural network, we need to ensure all sequences have the same length. We can do this by padding the sequences using `tf.keras.preprocessing.sequence.pad_sequences`.

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

maxlen = 256  # Or another appropriate length for your task

x_train_padded = pad_sequences(x_train, maxlen=maxlen)
x_test_padded = pad_sequences(x_test, maxlen=maxlen)

print("Padded training data shape:", x_train_padded.shape)
print("Padded testing data shape:", x_test_padded.shape)
```

```python
x_train = matrix_seq(x_train)
x_test = matrix_seq(x_test)
```

```python
x_train[0]
```

```python
y_train = np.asarray(y_train).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16,activation='relu',input_shape = (5000,)),
    tf.keras.layers.Dense(units=16,activation='relu',input_shape = (5000,)),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
    ])
model.summary()
```

```python
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
x_val = x_train[:10000]
x_train_1 = x_train[10000:]
```

```python
y_val = y_train[:10000]
y_train_1 = y_train[10000:]
```

```python
h = model.fit(x_train_1, y_train_1, epochs = 50, batch_size=512, validation_data = (x_val, y_val))
```
