```python
import numpy as np
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
X = np.arange(0,12,0.5)
```

```python
y = np.sin(X) + np.random.randint(1,10)
```

```python
plt.plot(X,y)
plt.show()
```

```python
clf = tree.DecisionTreeRegressor(random_state=0, max_depth = 10, min_samples_leaf =5)
```

```python
X.reshape(1,-1)
```

```python
clf.fit(X.reshape(len(X),1),y)
```

```python
clf.predict([[2.5]])
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
y_pred = cross_val_predict(clf, X.reshape(len(X),1), y, method = "predict")
```

```python
plt.plot(X,y)
plt.plot(X,y_pred)
plt.show()
```
