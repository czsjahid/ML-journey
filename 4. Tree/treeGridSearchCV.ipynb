```python
from sklearn.datasets import make_moons
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_data, y_data = make_moons( n_samples = 1000, noise = 0.5, random_state = 42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_data,y_data , train_size = 0.9, random_state =42)
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = { "min_samples_leaf": list(range(1,2000))}
```

```python
tree_GridSearchCV = GridSearchCV( clf, params, n_jobs =1)
```

```python
tree_GridSearchCV.fit( X_train, y_train)
```

```python
tree_GridSearchCV.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import cross_val_predict
```

```python
y_pred = cross_val_predict( tree_GridSearchCV, X_train, y_train, method = "predict")
```

```python
confusion_matrix(y_train, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracy = cross_val_score( tree_GridSearchCV, X_train, y_train,scoring = "accuracy")
```
